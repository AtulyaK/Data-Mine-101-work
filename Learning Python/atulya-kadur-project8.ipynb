```python
station_id_temp = 1
my_year = 2020
my_results = myDF[(myDF['station_id'] == station_id_temp) & (myDF['year'] == my_year)]
my_results.boxplot(column = 'wind_speed_mph', by = 'month')
plt.title('box plot of wind speed by year')
plt.xlabel('month')
plt.ylabel('wind speed ')
```

```python
def boxplot(my_year, station_id_temp):
    my_results = myDF[(myDF['station_id'] == station_id_temp) & (myDF['year'] == my_year)]
    my_results.boxplot(column = 'wind_speed_mph', by = 'month')
boxplot(2019, 3)
plt.title('box plot of wind speed by year')
plt.xlabel('month')
plt.ylabel('wind speed ')
```
