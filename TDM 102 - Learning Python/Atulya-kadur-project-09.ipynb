## Question 4

```python
#question 4 answer
def temperature_and_location(dataframe):
    dataframe_copy = dataframe.dropna(subset=['temperature'])
    dataframe_copy['location'] = dataframe_copy['latitude'].astype(str) + ', ' + dataframe_copy['longitude'].astype(str)
    return (dataframe_copy.groupby('location'))['temperature'].mean()
temperature_and_location(myDF_copy)
```
